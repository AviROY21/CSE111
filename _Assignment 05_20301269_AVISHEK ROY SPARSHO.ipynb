```python
#TASK-4

class Color:
    def __init__(self,clr):
        self.clr=clr
    def __add__(self, other):
        if (self.clr=='yellow' and other.clr == 'red') or (self.clr == 'red' and other.clr =='yellow'):
            return Color('Orange')
        elif (self.clr=='red' and other.clr == 'blue') or (self.clr=='blue' and other.clr == 'red'):
            return Color('Violet')
        elif (self.clr=='yellow' and other.clr == 'blue') or (self.clr=='blue' and other.clr == 'yellow'):
            return Color('Green')

C1=Color(input('First Color: ').lower())

C2=Color(input('Second Color: ').lower())

C3= C1 + C2

print('Color formed:',C3.clr)

```

```python
#TASK-8

class Coordinates:
  def __init__(self,c1,c2):
    self.c1=c1
    self.c2=c2

  def __sub__(self,other):
    c1=self.c1-other.c1
    c2=self.c2-other.c2
    return Coordinates(c1,c2)

  def __mul__(self,other):
    c1 = self.c1*other.c1
    c2 = self.c2*other.c2
    return Coordinates(c1,c2)

  def __eq__(self,other):
    if self.c1==other.c1 and self.c2==other.c2:
      return 'The calculated coordinates are the same.'
    else:
      return 'The calculated coordinates are NOT the same.'
  def detail(self):
    return (self.c1,self.c2)
p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))
p4 = p1 - p2
print(p4.detail())
p5 = p1 * p2
print(p5.detail())
point_check = (p4 == p5)
print(point_check)
```
